1. define

```python
# Python Quiz Game using Functions and Loops

def run_quiz(questions):
    score = 0
    print("üß† Welcome to the Python Quiz Game!\n")

    for idx, q in enumerate(questions):
        print(f"Q{idx+1}. {q['question']}")
        for i, option in enumerate(q['options']):
            print(f"{i+1}. {option}")
        try:
            answer = int(input("Enter your answer (1-4): "))
            if q['options'][answer - 1].lower() == q['answer'].lower():
                print("‚úÖ Correct!\n")
                score += 1
            else:
                print(f"‚ùå Wrong! Correct Answer: {q['answer']}\n")
        except:
            print("‚ùó Invalid input. Skipping question.\n")

    print(f"üéâ Quiz Completed! You scored {score} out of {len(questions)}.")

# List of quiz questions
quiz_questions = [
    {
        "question": "What is the keyword to define a function in Python?",
        "options": ["function", "def", "define", "func"],
        "answer": "def"
    },
    {
        "question": "Which of these is a valid variable name?",
        "options": ["2var", "_name", "user-name", "None"],
        "answer": "_name"
    },
    {
        "question": "Which function is used to take input from the user?",
        "options": ["print()", "input()", "scanf()", "read()"],
        "answer": "input()"
    },
    {
        "question": "What does 'len()' do in Python?",
        "options": ["Adds numbers", "Returns length", "Creates a loop", "None"],
        "answer": "Returns length"
    },
    {
        "question": "Who created Python?",
        "options": ["James Gosling", "Guido van Rossum", "Dennis Ritchie", "Bjarne Stroustrup"],
        "answer": "Guido van Rossum"
    }
]

# Run the quiz
run_quiz(quiz_questions)
```

Activity 2: Word Frequency Counter (Mini NLP)

Skill: String, Loops, Dictionary
Description: Students will write a Python script on Colab to:

Accept a paragraph as input
Count the frequency of each word
Display sorted word-frequency pairs

Key Concepts:
input(), split(), for loop
dictionary for frequency count
sorted() function

Bonus: ignore punctuation, convert to lowercase

Sample Output:
yaml
Copy
Edit
Input:
"Python is easy. Python is powerful."

Output:
{'python': 2, 'is': 2, 'easy': 1, 'powerful': 1}
‚è±Ô∏è Duration: 45 minutes
‚úÖ Deliverable:
Colab notebook submission via GitHub or shared link. Optional: Display results as a bar chart using matplotlib (for advanced students).



```python
# Word Frequency Counter in Python

def clean_text(text):
    # Remove punctuation and convert to lowercase
    import string
    for char in string.punctuation:
        text = text.replace(char, "")
    return text.lower()

def word_frequency_counter():
    print("üìò Word Frequency Counter")
    paragraph = input("Enter a paragraph:\n")

    # Clean and split the paragraph
    cleaned = clean_text(paragraph)
    words = cleaned.split()

    # Count word frequencies
    word_freq = {}
    for word in words:
        word_freq[word] = word_freq.get(word, 0) + 1

    # Display results
    print("\nüìä Word Frequency Table:\n")
    for word, count in sorted(word_freq.items(), key=lambda x: x[1], reverse=True):
        print(f"{word:<15} : {count}")

# Run the counter
word_frequency_counter()
```
